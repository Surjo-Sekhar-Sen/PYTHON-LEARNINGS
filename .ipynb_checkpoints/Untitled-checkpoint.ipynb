```python
#CALL BY OBJECT REFERENCE
a = 4
print(id(a))
print(id(4)) #BOTH OF THEM HAVE SAME ID'S
#USUALLY THE ID WE SEE IS GIVEN AS A NUMBER..MOSTLY THEY ARE IN HEXADECIMAL FORM!!
print(hex(id(a)))
print(hex(id(4)))
```

```python
#ALIASING
a = 4
b = a
print(id(a))
print(id(b)) #BOTH OF THEM WILL POINT TO THE SAME LOCATION!!
c = b
d = c
print(id(c))
print(id(d)) #EVERYTHING WILL POINT TO THE SAME MEMORY ADDRESS!! AS ALL POINT TO 4!!
```

```python
#WHAT IF WE DELETE a THEN WILL b, c AND d BE GONE??
a = 4
b = a
c = b
d = c
del a
print(b)
print(id(b))
del b
print(c)
print(id(c))
del c
print(d)
print(id(d))
```

```python
#WHAT IF WE CHANGE THE VALUE OF a, THEN WOULD b BE POINTING TO THE SAME ADDRESS?
#YES
a = 4
b = a
print(a)
print(id(a))
print(b)
print(id(b))
a = 5
print(a)
print(id(a)) #HERE ID OF a WILL CHANGE AS IT'S NOW POINTING TO A DIFFERENT OBJECT!!
print(b)
print(id(b)) #THIS WILL STILL POINT TO THE SAME OBJECT BCOZ THE OBJECT REFERENCE DIDN'T
#CHANGE!!
```
