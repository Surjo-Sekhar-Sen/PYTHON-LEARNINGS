```python
#STRING FUNCTIONS
#COMMON FUNCTIONS WITH STRING GIVEN AS AN ARGUMENT!
a = "Hello"
print(len(a))
print(max(a)) #Gives according to ASCII values!!
print(min(a))
print(sorted(a)) #This sorts the string in ascending order of ASCII values! 
# and prints in list!
print(sorted(a, reverse=True)) #Reverse is a hidden parameter we can use to reverse the string!
```

```python
#FUNCTIONS WITH NO PARAMETERS!!
#CAPITALIZE FUNCTION
b = "surjo"
print(b.capitalize()) #REMEMBER THAT THIS IS TEMPORARY RESULT, 
# IT DOESN'T EFFECT THE ACTUAL STRING!
print(b)
```

```python
#TITLE FUNCTION
b = "it is raining"
b.title()
```

```python
#UPPER AND LOWER FUNCTIONS
b = "KoLkatA"
print(b.upper())
print(b.lower())
```

```python
#SWAPCASE FUNCTION
b = "KoLKatA" #swaps the upper case values to lower and vice-versa!!
print(b.swapcase())
```

```python
#COUNT FUNCTION -> TO COUNT THE FREQUENCY OF A CHARACTER OR SUB-STRING!!
print("it is raining".count("i"))
print("It is raining".count("I")) 
print("it is raining".count("ing")) #can also work with sub-strings!!
print("it is raining".count("f"))
```

```python
#FIND FUNCTION
print("surjo sekhar sen".find("u")) #finds the first most index where the value is present!
print("surjo sekhar sen".find("sen"))
print("surjo".find("t")) #If the value is not present then it returns -1 !!
```

```python
#INDEX FUNCTION
print("surjo".index("u")) #similar to find function
print("surjo".index("t")) #will throw an error!!..
# THIS IS THE DIFFERENCE BETWEEN find() AND index(), index gives error whereas find gives -1
```

```python
#STARTSWITH AND ENDSWITH FUNCTION
print("it is sunday today".endswith("day"))
print("it is sunday today".endswith("to"))
print("it is sunday today".startswith("It"))
print("it is sunday today".startswith("it"))
print("it is sunday today".startswith("sunday"))
```

```python
#VERY IMPORTANT!!
#FORMAT FUNCTION
print("Hello guys, my name is {} and i am learning {}.". format("Surjo", "Python"))
print("Hello myself {name} and i am {age} years old.".format(name="Surjo", age= 21))
print("I am learning {1} from {0} channel. He is teaching {2}.".format("campusx", "python", 
                                                                       "great"))
```

```python
#ISALPHA FUNCTION -> checks for alphabetical values!
print("FLAT".isalpha())
print("FLAT2%".isalpha())
```

```python
#ISALNUM FUNCTION -> checks for alpha numeric values!
print("FLAT20".isalnum())
print("20% OFF".isalnum())
```

```python
#ISDECIMAL FUNCTION
print("sss".isdecimal())
print("555".isdecimal())
```

```python
#ISDIGIT FUNCTION
#ISNUMERIC FUNCTION
print("2023819.5".isdigit()) #HERE DOT(.) GETS TREATED LIKE A CHARACTER!! THATSWHY ITS FALSE
print("2023819".isdigit()) 
print("8289479817.45".isnumeric())
print("828947981745".isnumeric())
```

```python
#ISLOWER AND ISUPPER FUNCTIONS
print("hello".islower())
print("Hello".isupper()) #checks whether all the charcters are upper or not!!
print("HELLO".islower())
print("HELLO".isupper())
print("Hello".isupper())
```

```python
#ISTITLE FUNCTION
print("Is it Raining?".istitle())
print("Is It Raining?".istitle())
```

```python
#ISSPACE FUNCTION
print("I like astrphysics".isspace())
print("".isspace())
print(" ".isspace())
print("\t\n".isspace())
```

```python
#ISIDENTIFIER FUNCTION
print("25surjo".isidentifier())
print("surjo_sen".isidentifier())
print("bdHUBCZHm8hhhf212i".isidentifier())
print("bhdh>msh&".isidentifier())
print("_____".isidentifier())
```

```python
#SPLIT FUNCTION
print("Hello everyone, this is split function".split())
print("Hello everyone, this is split function".split("this"))
print("Hello everyone, this is split function".split("s"))
print("Hello everyone, this is split function".split(","))
print("Hello everyone, this is split function".split("surjo")) #doesn't split!!
```

```python
#JOIN FUNCTION
print(" ".join(['Hello', 'everyone,', 'this', 'is', 'split', 'function']))
print("/".join(['Hello', 'everyone,', 'this', 'is', 'split', 'function']))
print("-".join(['Hello', 'everyone,', 'this', 'is', 'split', 'function']))
print("<>".join(['Hello', 'everyone,', 'this', 'is', 'split', 'function']))
```

```python
a = "Hello everyone, this is split function".split(",")
b = "".join(a).split()
print("/".join(b))
```

```python
#REPLACE FUNCTION
"Hi my name is Surjo".replace("Surjo","Sekhar")
```

```python
#STRIP FUNCTION -> IT STRIPS THE WHITESPACES!!
#THIS PROVES HELPFUL WHEN USER BY MISTAKE PUTS A LOT OF SPACES WHEN FILLING DATA!!
name = "               Surjo     "
name.strip()
```

```python
name = "     Surjo      Sekhar  Sen   " #IT ONLY REMOVES THE ENDS EXTRA SPACES!!
name.strip()
```

```python
name = "     Surjo      Sekhar  Sen   " 
b = name.strip().split()
print(" ".join(b))
```
