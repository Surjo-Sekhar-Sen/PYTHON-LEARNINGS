```python
#SETS
#CREATION
S = {} #this is not treated as a set!! instead treated as dictionary!!
print(type(S))
#FOR CREATING AND EMPTY SET! WE NEED TO USE THE set() function!
S = set()
print(type(S))
```

```python
#IMPORTANT
#SETS ARE UNORDERED AND CANNOT CONTAIN DUPLICATE ITEMS!!
S = {1, 1, 2, 3, 4, 4}
print(S)
S = {1, 2, 3, 4, 5} #sets are unordered and thats why we can't access elements by indexing!
print(S[0])
```

```python
#SET IS A MUTABLE DATA TYPE ITSELF, BUT CANNOT BE EDITED..ONCE A VALUE IS PUT, IT CANNOT BE
#EDITED..WHICH MEANS IT'S ITEMS ARE IMMUTABLE!!
S = {(1, 2), "HELLO"}
print(S) #THIS IS ALLOWED AS TUPLES AND STRINGS ARE IMMUTABLE!!
S = {[1,2],1, 2} #IF YOU TRY TO STORE A MUTABLE VALUE LIKE LIST THEN IT WILL THROW ERROR!
print(S)
```

```python
S = {(1, 2), (22, 3), "Surjo", 1, 2, 3.56, None, True} 
print(S) # SETS ARE HETEROGENOUS! BUT ARE UNORDERED..
#THIS ORDER IS DECIDED ACCORDING TO HASHING!!..AND HASHING HAPPENS ONLY FOR IMMUTABLE THINGS
```

```python
#ADDING
#SETS ARE MUTABLE BUT NOT EDITABLE
S = {1, 2, 3, 4}
S.add(5) #adds a single value!
print(S)

S = {1, 2, 3, 4, 6}
S.update([11, 23, 41], "surjo") #add multiple values!
print(S)
```

```python
#WHAT IF I CHANGE THE SET INTO A LIST AND THEN EDIT VALUES FROM IT
#AND CHANGE IT BACK TO SET!!
S = {1, 2, 3, 4, 5}
print(id(S))

L = list(S)
L[0] = 100
print(L)
S = set(L)
print(S)
print(id(S))
#WELL THIS DOESN'T WORK AS THEY WILL NOT POINT TO THE SAME ADDRESS!! KIND OF RE-ASSIGNING!
```

```python
#DELETING
#WE CAN DELETE WHOLE SET BY USING del 
S = {1, 2, 3, 4, 5}
del S
print(S) 
```

```python
#CLEAR
S = {1, 2, 3, 4}
S.clear() #Clears the set values and empties it!!
print(S)
```

```python
#REMOVE AND DISCARD
S = {1, 2, 3, 4}
S.remove(4) #throws error if can't find the item to delete!!
print(S)
S.discard(3) #doesn't throw if can't find the item!!
print(S)
```

```python
S = {1, 2, 3, 4}
S.discard(5)
print(S)
S.remove(6)
print(S)
```

```python
#POP
S = {1, 2, 3, 4}
S.pop() #pops the last item from the set!!BUT SET IS UNORDERED-> THE ITEM TO BE POPPED 
#IS DECIDED BY HASHING!!
print(S)
S.pop()
print(S)

S = {"hello", 2.34, None, True, 1, 21, "surjo"}
S.pop()
print(S)
S.pop()
print(S)
```

```python
#SET OPERATIONS
S1 = {1, 2, 3, 4} #SET CONCATENATION DOESN'T WORK!
S2 = {"hello", 44, 56} #SET MULTIPLICATION ALSO DOESN'T WORK!!
print(S1 + S2)
print(S1 * 2)
```

```python
#ITERATION LOOPING
S = {1, 2, 3, 4, "hello", (1, 2, 3)}
for i in S:
    print(i, end = " ")
```

```python
#MEMBERSHIP
print(2 in S)
print(5 in S)
print("surjo" not in S)
print((1,2) not in S)
```

```python
#FUNCTIONS ON SETS
S = {1, 2, 3, 4, 5}
T = {1, 3, 5, 31, 3.52, True}
print(len(S))
print(min(S))
print(max(S))
print(sum(S))
print(sorted(S))
print(sorted(S, reverse = True))
print(sorted(T))
print(sorted(T, reverse = True))
```

```python
#MATHEMATICAL OPERATIONS
S1 = {1, 2, 3}
S2 = {2, 3, 4, 5}
print(S1.union(S2))
print(S1.intersection(S2))
print(S1.difference(S2))
print(S1.symmetric_difference(S2))
print(S2.difference(S1))
print(S2.symmetric_difference(S1))

print(S1.isdisjoint(S2))
print(S1.issubset(S2))
print(S1.issuperset(S2))
```
