```python
#BUILT-IN FUNCTIONS
#PRINT
print("Hello Everyone!!")
```

```python
#INPUT
name = input("Enter your name")
```

```python
#TYPE
a = type(1)
b = type(2.5)
c = type([1,2,3,4])
d = type("hello")
e = type({1,4,57,3})
f = type((2,3,4,56))
g = type(None)
h = type(False)
print(a, b, c, d, e, f, g, h)
```

```python
#INT
print(int('4'), int(5.4), int(4.563224))
#FLOAT
print(float('3.56'), float(45))
#STR
print(str(23), str(2.34556))
#THERE ARE SIMILAR FUNCTIONS LIKE TUPLE, LIST, SET,ETC....
```

```python
#ABS -> Helps to find the absolute value or the positive value!!
print(abs(-4), abs(4))
```

```python
#POW
print(pow(1, 3), pow(2, 3), pow(3,-3), pow(4, 3))
```

```python
#MIN/MAX
a = min([2, 4, 12, 43, 24, 92])
b = max([23,4,12,56,7,90,14])
c = min("Kolkata")
d = max("Kolkata")
e = max("KOLKaTA") #a has more  ASCII value than the capital letters!!
print(a, b, c, d, e)
```

```python
#ROUND -> It round off the values to nearest integer!!
pi = 22/7
print(round(pi)) #Here it rounds to nearest integer!
print(round(pi, 2)) #Here it rounds with 2 decimal places as asked!!
```

```python
#DIVMOD -> Gives the quotient and remainder of two numbers (a, b) a divided by b!!
print(divmod(4, 2)) # (4 // 2 = 2, 4 % 2 = 0)
print(divmod(7, 3))
```

```python
#BIN/OCT/HEX
#For getting binary values of a number..bin()
#For Octal..oct()
#For Hexadecimal..hex()
print(bin(4), bin(23), oct(12), oct(10), hex(30), hex(23))
```

```python
#ID -> Helps to find the memory address of the variable!!
a = 3
b = 45
print(id(a), id(b))
```

```python
#ORD -> This is used to find the ASCII Code of any character!!
print(ord('a'), ord('C'), ord('Z'), ord('e'))
```

```python
#LEN
print(len("Kolkata"), len([1, 2, 3, 5, 12, 41, 24, 12, 7]))
```

```python
#SUM
sum([1, 2, 3, 12, 32, 10, 5])
```
