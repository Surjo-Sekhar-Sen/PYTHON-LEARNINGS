```python
#RECURSION
#LETS SAY MULTIPLICATION OPERATOR DIDN'T EXIST, SO HOW WOULD YOU MULTIPLY TWO
#NUMBERS AND GIVE OUTPUT!!
#HERE WE CAN DO LOOPING OR WE CAN ALSO USE RECURSION!!
```

```python
def multiply(a, b):
    result = 0
    for i in range(b):
        result = result + a

    return result
print(multiply(3, 4))
print(multiply(5, 9))
print(multiply(8, 7))
print(multiply(7, 3))
```
