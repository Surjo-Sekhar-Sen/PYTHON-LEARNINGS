```python
#MUTABILITY
#IT BASICALLY MEANS IF YOU ARE ABLE TO CHANGE THE VALUE STORED IN ANY MEMORY ADDRESS,
#WITHOUT EFFECTING THE MEMORY ADDRESS!!
a = "Hello"
print(a)
print(id(a))
a = a + "World"
print(a)
print(id(a)) #HERE WE SEE THE ID OF a HAS CHANGED!! BCOZ STRINGS ARE IMMUTABLE!!
print(id("Hello")) #THE id OF Hello and Initial a IS SAME!!
```

```python
T = (1, 2, 3)
print(T)
print(type(T))
print(id(T))
T = T + (4, 5)
print(T)
print(type(T)) #THE CLASS IS STILL TUPLE!!
print(id(T)) #BUT THE ID CHANGED BCOZ TUPLE IS IMMUTABLE!!
```

```python
#LIST - MUTABLE DATA  TYPES
L = [1, 2, 3]
print(L)
print(id(L))
L.append(4) #THIS MAKES CHANGES ON THE MEMORY ID!!
print(L)
print(id(L))
```

```python
#LIST APPEND AND ALL FUNCTIONS WORK BUT LIST CONCATENATION CHANGES THE MEMORY ID!!
L = [2, 3.9]
print(id(L))
L = L + [40, 32] #THIS DOES NOT CHANGE THE VALUE IN THE MEMORY ID INSTEAD MAKES A 
#DIFFERENT LIST!!
print(L)
print(id(L))
```

```python
#ALIASING BUG
L1 = [1, 2, 3]
L2 = L1
print(L1, L2)
print(id(L1), id(L2))
L2.append(4)
print(L1, L2) #HERE WE DIDN'T WANT TO CHANGE L1 BUT BCOZ OF ALIASING IT ACTUALLY CHANGES 
#THE SAME MEMORY ADDRESS!!
```

```python
L1 = [1, 2, 3]
L2 = L1[:] #BY DOING THIS, IT WILL COPY ITS REFERENCE AND HAVE DIFFERENT MEMORY ID!!
print(L1, L2)
print(id(L1), id(L2)) #ID'S ARE DIFFERENT!!
```

```python
#(Q). WILL THE LIST ELEMENT CHANGE IF THE LIST IS AN ELEMENT OF A TUPLE!??
T = (1, 2, 3, [4, 5])
print(T)
print(id(T))
print(T[-1])
print(T[-1][-1])
T[-1][-1] = 500
print(T)
print(id(T)) #WILL HAVE THE SAME MEMORY ID!!
```

```python
#(Q). WILL THE TUPLE ELEMENT CHANGE IF THE TUPLE IS AN ELEMENT OF A  LIST!??
L = [1, 2, 3, (4, 5)]
print(L)
print(id(L))
print(L[-1][-1])
L[-1][-1] = 500 #THIS WILL THROW AN ERROR!!
print(L)
print(id(L)) 
```

```python
L1 = [1, 2]
L2 = [3, 4]
T = (L1, L2) #TUPLE WITH TWO LIST ELEMENTS!!
print(T)
print(id(T))
T[0][0] = 100 #THIS DOESN'T CHANGE THE MEMEORY ADDRESS!!
print(T)
print(id(T))
```

```python
L1 = [1, 2]
L2 = [3, 4]
T = (L1, L2) #TUPLE WITH TWO LIST ELEMENTS!!
print(T)
print(id(T))
T[0] = T[0] + [6, 7] #THIS WILL THROW AN ERROR AS IT TRIES TO CHANGE THE ELEMENT OF TUPLE!
print(id(T))
```

```python
#WE CAN JUST USE THE copy() FUNCTION TO PREVENT THIS!! THIS CAN ALSO PREVENT THE PROBLEM OF
#ALIASING!!
a = [1, 2, 3]
b = a.copy()
b.append(4)
print(a, b)
```
