```python
#THESE ARE THE BASIC DATA TYPES IN PYTHON( int, float, bool, str, complex ) 
#integer(int)
print(5)
print(50000000000000000000000000000)
print(1e308) #this is the highest value that can be used for int datatype!!
print(1e309) #this will give infinity(inf) value as return!!
```

```python
#float(float)
print(2.7)
print(3.7877)
print(1.7e308) #this is the highest value that can be used for float datatype!!
print(1.7e309) #this will give infinity(inf) value as return !!
```

```python
#boolean(bool)
print(True)
print(False)
```

```python
#string(str)
print('Hello Everyone')
print("My name is Surjo")
print('''My name is Surjo Sekhar Sen''')
```

```python
#complex(complex)
print(1+2j)
print(2-4j)
```

```python
#NOW THERE ARE CONTAINER DATA TYPES IN PYTHON ( list, tuple, sets, dict )
#List
print([1,2,3,4,5])
```

```python
#tuple
print((1,2,3,4,5))
```

```python
#sets
print({1,2,3,4,5})
```

```python
#dictionary(dict)
print({"Name":"Surjo","Age":20,"Gender":"Male"}) #dictionary contains key-value pairs!!
```
