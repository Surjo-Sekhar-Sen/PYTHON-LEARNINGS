```python
#MUTABILITY
#IT BASICALLY MEANS IF YOU ARE ABLE TO CHANGE THE VALUE STORED IN ANY MEMORY ADDRESS,
#WITHOUT EFFECTING THE MEMORY ADDRESS!!
a = "Hello"
print(a)
print(id(a))
a = a + "World"
print(a)
print(id(a)) #HERE WE SEE THE ID OF a HAS CHANGED!! BCOZ STRINGS ARE IMMUTABLE!!
print(id("Hello")) #THE id OF Hello and Initial a IS SAME!!
```

```python
T = (1, 2, 3)
print(T)
print(type(T))
print(id(T))
T = T + (4, 5)
print(T)
print(type(T)) #THE CLASS IS STILL TUPLE!!
print(id(T)) #BUT THE ID CHANGED BCOZ TUPLE IS IMMUTABLE!!
```

```python
#LIST - MUTABLE DATA  TYPES
L = [1, 2, 3]
print(L)
print(id(L))
L.append(4) #THIS MAKES CHANGES ON THE MEMORY ID!!
print(L)
print(id(L))
```

```python
L = [2, 3.9]
print(id(L))
L = L + [40, 32] #THIS DOES NOT CHANGE THE VALUE IN THE MEMORY ID INSTEAD MAKES A 
#DIFFERENT LIST!!
print(L)
print(id(L))
```
