```python
def is_even(number): #this is function declaration and definition with a parameter!
    """Input: an integer number,
    This function will tell us whether a number is even or odd!!
    Output: Odd/Even""" #This is docstring!
    
    if(number % 2 == 0): #this is the inside code of the function!
        return "Even!"
    else:
        return "Odd!"
print(is_even(3)) #function calling!!
print(is_even(4))
```
