```python
#QUESTION PROBLEM 1
#IF I AM GIVEN A STRING:- "hello how are you"
#NOW  IF I WANT TO CHANGE THE FIRST LETTER OF THE WORDS TO CAPITAL LETTER!!
#WE USUALLY WOULD HAVE USED .title() function BUT DON'T USE THAT!!
#CREATE A CUSTOM LOGIC TO DO THIS!!
```

```python
string = "hello how are you"
strings = (string.split())
new_string = []
for word in strings:
    new_string.append(word.capitalize())
print(" ".join(new_string))
```

```python
#QUESTION PROBLEM 2
#IF I AM GIVEN AN EMAIL-ID LIKE: abcd@gmail.com
#NOW EXTRACT THE THINGS WHICH COME BEFORE @!!
```

```python
email = input("Enter your email")
pos = email.find('@')
print(pos)
for val in email[0:pos]:
    print(val, end="")
```

```python
#QUESTION PROBLEM 3
#IF I AM GIVEN A LIST WITH DUPLICATED ITEMS LIKE: [1, 1, 2, 2, 3, 3, 4, ,4]
#GIVE THE LIST WITH NO DUPLICATE ITEMS!!
```

```python
L = [1, 1, 2, 2, 3, 3, 4, 4]
L1 = []
for l in L:
    if l not in L1:
        L1.append(l)
print(L1)
```
