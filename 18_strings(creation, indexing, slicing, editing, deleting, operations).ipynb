```python
#CREATING STRINGS
a = "Hello" #Double inverted commas
print(a)

b = 'Hello' #Single inverted commas
print(b)

c = '''Hello my name is Surjo
I am learning python!!''' #Triple inverted commas (USED FOR MULTI-LINE STRINGS!!)
print(c)

d = """Hello""" #Triple double-inverted commas
print(d)

e = str(23456)
print(e)
```

```python
#INDEXING
a = "Hello"
print(a[0], a[1], a[2], a[4])
```

```python
#There are two types of indexing in python!
#POSITIVE INDEXING
b = "Surjo"
print(b[2], b[4])
```

```python
#NEGATIVE INDEXING
c = "Sekhar"
print(c[-1], c[-2], c[-3], c[-4])
```

```python
#BY USING NEGATIVE INDEXING WE CAN GET THE LAST CHARACTER OF ANY STRING EVEN IF
#WE DON'T KNOW THE FULL LENGTH OF THE STRING!!
```

```python
#SLICING
a = "Hello World"
print(a[0:2]) #a[start:end], Here the end is excluded! So till (end-1)!!
```

```python
print(a[-5:-1]) #Negative indexing also works, but end is excluded!
```

```python
b = "Hello World"
print(b[2:7]) #Space is also a character!!
```

```python
print(b[2:3]) #If we have been told to print the specific character only then we can do this!
```

```python
c = "Surjo Sekhar Sen"
print(c[2:8:2]) #c[start:end:step] Here this step means the gaps to jump!!
```

```python
print(c[2:8:-1]) #This won't give anything, as we used positive indexing 
#if we go backwards from the start that is 2nd pos by -1 and keep going, we won't reach 8th!
#but if we use negative indexing and positive step then it will work!!
```

```python
print(c[8:2:-1]) #We need to keep this in mind!! REVERSING WORKS ONLY WHEN WE CAN MOVE FROM
#RIGHT TO LEFT, SO MAKE SURE TO GIVE THE START AND END CAREFULLY!!
```

```python
print(c[--8:-1:2])
```

```python
print(c[-1:-8:-2]) #If we want negative step then start with the end of the string
#as we want to go backwards!!
```

```python
d = "Hello Everyone"
print(d[2:]) #This means from 2 then till the end!
print(d[:6]) #This means from start till (6-1)!
print(d[:]) #This means from start to end, total!!
```

```python
#VERY IMPORTANT
#FOR REVERSING A STRING!!
e = "Surjo"
print(e[::-1])
```

```python
#EDITING STRINGS
a = "Surjo"
a[2] = 't'
print(a)
```

```python
#This error comes because we cannot edit the values inside a string!!
#THIS IS BECAUSE STRINGS ARE IMMUTABLE(NOT CHANGABLE)!!
#ALSO WE CANNOT ADD CHARACTERS TO THE EXISTING STRING LIKE:
a[5] = 'Y'
print(a)
```

```python
#BUT WE CAN REASSIGN THE WHOLE STRING VALUE!!
a = "Sekhar"
print(a)
```

```python
"Mumbai" > "Pune" #This comaprison works lexiographically!!
#IT DOESN'T WORK ON THE BASIS OF ASCII OR UNICODE OR WHICH HAS MORE LENGTH!!
#LEXIOGRAPHICALLY MEANS THE WORD WHICH COMES LATER IN DICTIONARY, THAT WORD WILL BE GREATER!
```

```python
print("Goa" < "Kolkata")
print("GOA" < "kolkata")
```

```python
#LOGICAL OPERATIONS
#IN PYTHON EMPTY STRINGS "" ARE TREATED AS FALSE!
#WHEREAS NON-EMPTY STRINGS ARE TREATED AS TRUE!!
print("" or "surjo") # 0 or 1 -> 1 (surjo gets printed as it is 1)
print("" and "surjo") # 0 and 1 -> 0 (empty "" string)
print("sen" and "sekhar") # 1 and 1 -> 1(sekhar confirms the answer can be 1 
#so sekhar gets printed)
print("" and "") # 0 and 0 -> 0 (empty "" string)
print("sen" or "surjo") # 1 or 1 -> 1 (sen confirms that 1 is present, 
#so need to look for any other 1)
```

```python
print(not "sen")
print(not "")
```

```python
#LOOPS IN STRINGS
c = "Hello World"
for i in c:
    print(i, end = "")
```

```python
for i in c[2:9:2]:
    print(i, end = "")
```

```python
for i in c[::-1]:
    print(i, end = "")
```
