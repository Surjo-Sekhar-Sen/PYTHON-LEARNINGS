```python
#LITERALS
#numeric literals
a=10 #decimal format
b=0b1010 #binary format
c=0o12 #octal format
d=0xa #hexadecimal format
print(a, b, c, d)
```

```python
#string literals
print('Hello') #single quote
print("Everyone") #double quote
print('''My name is surjo 
and i am fine!''') #triple quote or docstrings
print(u'\U0001f600') #uncicode, helps to print emojis!!
print(r'C:\name\path') #raw string like path or address, it ignores the escape sequences like \n or \t!!
```

```python
#boolean literals
a=10+True #True is treated as 1 in python and here we also see implicit type conversion from bool to int!!
b=5-True
c=9+False #False is treated as 0 in python!!
d=72-False
print(a, b, c, d)
```

```python
#Numeric literals: float and complex
a=1.5
b=1.5e2
c=1.5e-3
print(a, b, c)
d=2+3j
e=5j
print(d, e)
```

```python
#None literal
#if we want to just declare a variable in python, it's not defining and initialising that we normally do in python!!
#for just declaring it we still have to inialise it with some value..we can do it with None!!
k=None
print(k)
```
