```python
#ACCESSING LISTS
#ACCESSING IN 1D LIST
L = [1, 2, 3, 4]
print(L[0], L[2]) #THIS INDEXING AND SLICING WORKS SIMILAR TO STRINGS!!
print(L[1:3])
print(L[-4:-1])
print(L[-1:-4:-1])
```

```python
#ACCESSING IN 2D LIST
L = [1, 2, 3, [4, 5, 6, 7]]
print(L[3])
print(L[3][0], L[3][1])
print(L[3][0: 2])
print(L[3][1: 3])
print(L[3][::2])
print(L[3][-1:-4:-2])
```

```python
#ACCESSING IN 3D LIST
L = [[1, [2, 8]],[[3, 9, 10, 11], 5, [13, 14]],[6, [16,4]]]
print(L[0], L[0][1][1], L[1][0][3])
print(L[2][0:1])
print(L[1][0][0:3])
print(L[1][::-1][0:2][0][1]) #At first we go into the 1st index then we reverse the elements
#then in the reversed list, we go from 0 to 1st index and in that we go to 0th which [13,14]
#and then we select its 1st index!!
```

```python
#EDITING IN LIST
L = [1, 2, 3, 4, 5]
L[0] = 100 #lists are mutable!! So they can be edited unlike strings!!
L[-1] = 500
print(L)
L[1:3] = [200,300]
print(L)
```

```python
#ADDING IN LIST
L = [10, 20, 30]
L.append(40) #append will add in the last!!
print(L)
L.append("hello") #string can also be appended as lists are  heterogenous!!
print(L)
L.append([5,6]) #append always adds only one value in the last..so if you try to add 
#multiple values then it will be taken as a single value!
print(L)
L.extend([50,60,70]) #extend is used to add multiple values in thee last!!
print(L)
L.extend("world") #this won't be added as a single value as extend always treats the given
#thing as multiple values!
print(L)
L.insert(2,"surjo") #this helps to add element in a specific location!
print(L)
```

```python
L = [1, 2, 3, 4]
L[0:2] = "surjo"
print(L)
print(L[0:6])
```

```python
#DELETING IN LIST
L = [1, 2, 3, 4, 5, 6, 7, 8, 9, "hello", "surjo",[20,30]]
del L[1] #del is not written like .del()..deletes the value or sliced values!!
print(L)
del L[4:7]
print(L)
```

```python
L = [1, 3, 4, 5, 9, 'hello', 'surjo', [20, 30]]
L.remove(4)
L.remove("surjo") #only takes single argument!!
print(L)
L.pop() #removes the last element only!!
print(L)
L.clear() #makes the list empty!!
print(L)
```

```python
#LIST CONCATENATION AND MULTIPLICATION
L1 = [1, 2, 3, 4]
L2 = [5, 6, 7, 8]
print(L1 + L2) #list gets concatenated together!!
print(L1 * 3)
```

```python
#MEMBERSHIP OPERATION
L1 = [1, 2, 3, [4,5]]
print(1 in L1)
print(4 in L1)
```

```python
#LOOPING IN LIST!! (IMPORTANT)
L1 = [1, 2, 3, 4, 5]
for i in L1:
    print(i, end = " ")
print("")  
L2 = [1, 2, 3, [4, 5]]
for i in L2:
    print(i , end = " ")
```

```python
#IMPORTANT LOOPING ACCESSING ALL ELEMENTS IN 1D, 2D, 3D MIXED LIST!!
L = [1, [2, [3,4]], 5, [6,[7,8]]]
for i in L:
    if isinstance(i, list):
        for j in i:
            if isinstance(j, list):
                for k in j:
                    print(k, end = " ")
            else:
                print(j, end = " ")
    else:
        print(i, end = " ")
```

```python
#FUNCTIONS AND LIST
L = [1, 2, 3, 4]
print(len(L))
print(min(L)) 
print(max(L))
L1 = ["hello","surjo"]
print(min(L1))
L1.append(2)
print(L1)
#print(min(L1)) #this doesn't work as string and integer cannot be compared!!
print(L.index(3))
```

```python
#SORT() AND SORTED()
L = [1, 3, 2, 5, 4, 6]
print(sorted(L)) #sorted() is a temporary operation!!
print(L)
print(sorted(L, reverse = True))
L.sort() #sort() is a permanent operation!! AND IT IS WRITTEN IN A VERY DIFFERENT MANNER!!
print(L)
L.sort(reverse = True)
print(L)
```
