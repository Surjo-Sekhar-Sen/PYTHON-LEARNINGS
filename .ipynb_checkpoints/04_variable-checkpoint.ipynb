```python
#We see that Surjo isn't getting printed instead 4 got printed!
#THIS FEATURE OF PYTHON IS CALLED: DYNAMIC TYPING!!(Which means that we don't need to specify datatype to the variable)
name=7.983
print(name)
```

```python
#Opposite to this is called STATIC TYPING!!( LIKE IN C, C++ TYPE LANGUAGES WE SPECIFY THE DATA TYPE OF VARIABLE)
```

```python
#PLUS THE FEATURE WHERE WE CAN SEE NO SPECIFIC DATATYPE IS ASSOCIATED WITH THE VARIABLE AND CAN BE CHANGED DYNAMICALLY 
#AND WITHOUT ANY TYPE CASTING OR ANYTHING, THIS IS CALLED AS: DYNAMIC BINDING!!
#IN C TYPE LANGUAGES IF WE WANT TO CHANGE THE TYPE THEN WE HAVE TO TYPECAST IT THEN IT CAN HAPPEN!! 
```

```python
#Ways of initialising variable:
#BY SEPERATING THEM WITH SEMI-COLONS!
a=3;b=4;c=5
print(a)
print(b)
print(c)
```

```python
#BY JUST WRITING THEM WITH COMMAS AND GIVING VALUES IN THE SAME STYLE!
a,b,c=6,7,8
print(a)
print(b)
print(c)
```

```python
#FOR EQUAL VALUES JUST GIVE THEM LIKE THIS!
a=b=c=9
print(a)
print(b)
print(c)
```
