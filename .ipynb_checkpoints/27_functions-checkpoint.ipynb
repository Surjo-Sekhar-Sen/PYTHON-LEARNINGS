```python
def is_even(number): #this is function declaration and definition with a parameter!
    """Input: an integer number,
    This function will tell us whether a number is even or odd!!
    Output: Odd/Even""" #This is docstring!
    
    if(number % 2 == 0): #this is the inside code of the function!
        return "Even!"
    else:
        return "Odd!"
print(is_even(3)) #function calling!!
print(is_even(4))
```

```python
#IF WE ARE WRITING A FUNCTION, THE CODE INSIDE MUST BE WRITTEN IN SUCH A WAY THAT IT CAN
#HANDLE ANY ERROR OR WRONG INPUT!!
#LIKE IN THIS FUNCTION:- WE MUST ADD THE THING WHERE IF A PERSON GIVES WRONG INPUT THEN
#WHAT TO DO!!
def is_even(number): #this is function declaration and definition with a parameter!
    """Input: an integer number,
    This function will tell us whether a number is even or odd!!
    Output: Odd/Even""" #This is docstring!
    if(type(number) == int): #OR WE CAN USE if(isintance(number, int)):
        if(number % 2 == 0): #this is the inside code of the function!
            return "Even!"
        else:
            return "Odd!"
    else:
        return "Not Allowed!!"
print(is_even(24))
print(is_even(53))
print(is_even("Hello"))
```

```python
#TYPES OF ARGUMENTS
#POSITIONAL ARGUMENTS
#THE WAY/ORDER YOU GIVE PARAMETERS, THE SAME ORDER IS FOLLOWED FOR ARGUMENTS!!
def power(a, b): #NORMAL PARAMETERS
    return a**b

print(power(2, 3)) #NORMAL ARGUMENTS 
print(power(3, 2))
```

```python
#DEFAULT ARGUMENTS
def power(a = 1, b = 1):
    return a**b

print(power(2, 3))
print(power(3)) #TREATS b AS 1
print(power(2))
```

```python
#KEYWORD ARGUMENTS-> WE DON'T HAVE TO FOLLOW THE ORDER IF WE DON'T KNOW IT!!
def power(a = 1, b = 1):
    return a**b

print(power(b = 2, a = 3))
```

```python
#ARBITRARY ARGUMENTS-> HOW TO GIVE MULTIPLE INPUTS LIKE IN print() WE CAN MULTIPLE INPUTS 
#AS MANY AS WANT!!
def flexi(*number): #WE SEE *args AND *kwargs LIKE PARAMETERS!! THIS MEANS WE ARE ALLOWED
    #TO GIVE MULTIPLE INPUTS IN IT!!
    product = 1
    print(number) #THIS BASICALLY CREATES A TUPLE!!
    print(type(number))
    for i in number:
        product = product*i
    return product
print(flexi(1,2,3,4,5))
```

```python
#GLOBAL AND LOCAL SCOPING!!
def g(y):
    print(x) #TAKES THE GLOBAL VARIABLE x PRESENT!!
    print(x+1)

x = 5
g(x)
print(x) #DOES NOT CHANGE THE ACTUAL GLOBAL VARIABLE VALUE!
```

```python
#GLOBAL AND LOCAL SCOPING!!
def funct(y):
    x = x + 1
x = 5
funct(x)
print(x)
```

```python
#GLOBAL AND LOCAL SCOPING
#WE CAN EXPLICITELY USE GLOBAL VARIABLE LOCALLY!!
def funct(y):
    global x #BY USING THE KEYWORD global WE CAN DO THIS!!
    x = x + 1
x = 5
funct(x)
print(x)
```

```python
#EXAMPLE TO SEE WHETHER THERE ARE CHANGES IN GLOBAL VARIABLE BCOZ OF LOCAL CHANGES!
def f(x):
    x = x+1
    print(id(x))
    print("x inside:", x)
    return x
x = 3
print(id(x))
z = f(x)
print("z outside:", z)
print("x outside:", x)
```

```python
#NESTED FUNCTIONS
def f():
    print("Inside f")
    def g():
        print("Inside g")
    g()    
f()
```

```python
def f():
    print("Inside f")
    def g():
        print("Inside g")
    g()    
g() #THIS WON'T WORK AS g() IS PRESENT INSIDE THE f() FUNCTION WHICH IS KINDOF HIDDEN 
#FROM THE MAIN GLOBAL WORLD!!
#THIS IS DONE BY A LOT OF PROGRAMMERS WHERE THEY HIDE THEIR CODE BY DOING THIS!!
```

```python
#EXAMPLE CODE OF SCOPING AND NESTED FUNCTIONS!!
def g(x):
    def h(x):
        x = x + 1
    x = x + 1
    print("x inside:", x)
    a = h(x)
    print(a)
    return x

x = 3
z = g(x)
print("x outside:", x)
print("z outside:", z)
```

```python
#FUNCTIONS AS OBJECTS
def f(num):
    return num**2
print(f(2))
print(f(4))

x = f #STRONG THE DECLARATION OF THE FUNCTION f IN x VARIABLE!!
print(x(5)) #THIS ALLOWS IT TO BE USED IN A FUNCTION WAY!!
print(type(x)) #HERE WE SEE THE TYPE AS FUNCTION CLASS AS FUNCTION ITSELF IS AN OBJECT!!
```

```python
print(f(9))
del f #HERE WE DELETE THE MAIN f FUNCTION!
print(x(6)) #BUT THE OBJECT REFERENCE FOR x IS STILL THERE THATSWHY x STILL WORKS!!
```

```python
#WE CAN PASS FUNCTION AS AN ARGUMENT TOO!!!!
def a():
    print("Inside a")
def b(z):
    print("Inside b")
    return z()
print(b(a)) #HERE WE GIVE a AS AN ARGUMENT TO FUNCTION b(z), SO z = a, BUT a  IS A FUNCTION
#AND THATSWHY ITS DEFINITION GOES INSIDE IT AND HENCE WE CAN CALL z()!!
```

```python
#WE CAN RETURN A FUNCTION TOO AS A VALUE!!!!
def a():
    def x(a, b):
        return a + b
    return x #HERE x IS A FUNCTION WHICH WE ARE RETURNING TO a()!!
value = a()(5, 4) #HERE WHEN WE CALL a() IT RETURNS x, WHICH MAKES value = x(5, 4)!!
print(value)
```
