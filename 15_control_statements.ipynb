```python
#CONTROL STATEMENTS
#BREAK
for i in range(1, 11):
    if i == 5:
        break
    print(i)
```

```python
#break is mainly used for breaking out from the loop..
#In real life applications we can use break when using linear search to find something
#in a database!!
```

```python
#CONTINUE
for i in range(1, 11):
    if i == 5:
        continue
    print(i)
```

```python
#Here we see 5 is not getting printed..
#Continue is used for this, it is used to skip the code after it and run the loop again!!
#In real life, if we are building a website and we are showing something out of stock
#we can use continue for those elements and continue showing the in stock items!!
```
