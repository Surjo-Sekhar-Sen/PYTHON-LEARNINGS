```python
#CALL BY OBJECT REFERENCE
a = 4
print(id(a))
print(id(4)) #BOTH OF THEM HAVE SAME ID'S
#USUALLY THE ID WE SEE IS GIVEN AS A NUMBER..MOSTLY THEY ARE IN HEXADECIMAL FORM!!
print(hex(id(a)))
print(hex(id(4)))
```

```python
#ALIASING
a = 4
b = a
print(id(a))
print(id(b)) #BOTH OF THEM WILL POINT TO THE SAME LOCATION!!
c = b
d = c
print(id(c))
print(id(d)) #EVERYTHING WILL POINT TO THE SAME MEMORY ADDRESS!! AS ALL POINT TO 4!!
```

```python
#WHAT IF WE DELETE a THEN WILL b, c AND d BE GONE??
a = 4
b = a
c = b
d = c
del a
print(b)
print(id(b))
del b
print(c)
print(id(c))
del c
print(d)
print(id(d))
```

```python
#WHAT IF WE CHANGE THE VALUE OF a, THEN WOULD b BE POINTING TO THE SAME ADDRESS?
#YES
a = 4
b = a
print(a)
print(id(a))
print(b)
print(id(b))
a = 5
print(a)
print(id(a)) #HERE ID OF a WILL CHANGE AS IT'S NOW POINTING TO A DIFFERENT OBJECT!!
print(b)
print(id(b)) #THIS WILL STILL POINT TO THE SAME OBJECT BCOZ THE OBJECT REFERENCE DIDN'T
#CHANGE!!
```

```python
#ONLY SOME ARE INTERNING..THAT MEANS INITIALLY ONLY SOME VALUES HAVE RESERVED MEMORY ID
#THE REST ARE NORMAL OBJECTS!! WHICH GET ALLOCATED ONCE OBJECT REFERENCE IS TO BE MADE!!
#THIS IS DONE BY THE PYTHON WHERE IT FINDS AN EMPTY SPACE IN THE MEMORY!!
print(id(1000))
print(id(10))
print(id(1))
print(id(3))
print(id(7))
print(id(70))
print(id(26))
print(id(2600))
```

```python
#WEIRD STUFF IN PYTHON 1
#ALIASING AND NOT ALIASING WITH CONTEXT OF NORMAL OBJECTS AND IMMORTAL OBJECTS!!
#ALIASING HAPPENS WHEN WE EXPLICITLY MAKE A VARIABLE EQUAL TO THE OTHER!!
a = 4
b = a #THIS IS ALIASING
print(id(a), id(b))

a = -5
b = -5 #THIS IS NOT ALIASING! THIS WILL BECOME EASIER TO UNDERSTAND WITH NORMAL OBJECTS!
print(id(a), id(b)) 

a = 257 
b = 257 #WE SEE THAT EVEN WHEN THEY ARE POINTING TO THE SAME VALUE, ID'S ARE STILL DIFF!!
print(id(a), id(b)) #THIS IS BCOZ IN PYTHON FROM -5 TO 256 THE VALUES ARE IMMORTAL OBJECTS
#WHICH MEANS THEY ALREADY HAVE SOME MEMORY ID ASSIGNED TO THEM AS THEY ARE ASSUMED
# TO BE FREQUENTLY USED!!
```

```python
#WEIRD STUFF IN PYTHON 2
import sys
a = 717
b = a
c = b
print(id(717), id(a), id(b), id(c))
print(sys.getrefcount(a))
d = 717
print(id(d)) #IT HAS THE SAME ID BUT IF YOU CHECK THE refcount NOW..IT WILL STILL BE THE 
#SAME
print(sys.getrefcount(a)) #IT STILL PRINTS 5..IT IS BCOZ d IS NOT ALIASED!!
#IF I ALIAS IT THEN THE refcount WILL INCREASE!!
d = c
print(sys.getrefcount(a))
```

```python
#WEIRD STUFF IN PYTHON 3
a = "surjo"
b = "surjo"
print(id(a), id(b)) #THIS GIVES SAME ID!

a = "surjo sekhar sen"
b = "surjo sekhar sen"
print(id(a), id(b)) #THIS COMES DIFFERENT!

a = "surjo_sekhar_sen"
b = "surjo_sekhar_sen"
print(id(a), id(b)) #THIS COMES SAME!
#THIS HAPPENS BCOZ OF VALID IDENTIFIER NAMES IN PYTHON!!
#IF THE VALUE WE ARE GIVING IS A VALID IDENTIFIER THEN IT GIVES SAME ID IF DIFF VARIABLES
#ARE GIVEN SAME VALUE!
#BUT IF NOT A VALID IDENTIFIER THEN THEY MAY BE GIVEN DIFF ID'S!!
```

```python
#ADDRESS ID IN A LIST
#IF WE CREATE A LIST, WE USUALLY ASSUME THE ADDRESS OF THE BLOCKS WILL BE CONTINUOS LIKE 
#ARRAYS IN C OR C++..BUT IT'S NOT..IT DOESN'T WORK THAT WAY!!
L = [1, 2, 3]
print(id(L)) #THIS L WILL HAVE SOME MEMORY ID SEPERATE IN THE MEMORY!!
#WHEN WE TRY TO GET THE ID OF THE ELEMENTS FROM THE LIST!
print(id(L[0]), id(1)) #HERE THE VALUE IN THE 1ST INDEX IS 1..SO THE ID OF THAT INDEX WILL
#HAVE ID SAME AS 1..
print(id(L[1]), id(2)) #THE 1ST INDEX ELEMENT IS 2..SO IT WILL HAVE 2'S MEMORY ID..
#THIS MAKES THE ID OF THE LIST INDICES DISCONTINUOS!!
```
