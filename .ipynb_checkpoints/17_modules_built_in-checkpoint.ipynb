```python
print(math.ceil(5.7)) #ceil function in math module will give the next integer value!!
print(math.ceil(5.3)) #THIS IS NOT round() function!!
```

```python
#RANDOM MODULE
import random
```

```python
print(random.randint(1, 100)) #GIVES RANDOM VALUES IN BETWEEN THE RANGE!!
print(random.randint(1, 100))
```

```python
a = [1, 2, 3, 4, 5]
random.shuffle(a) #SHUFFLES THE ITEMS GIVEN..LIKE CARD SHUFFLING!!
print(a)
random.shuffle(a)
print(a)
```
