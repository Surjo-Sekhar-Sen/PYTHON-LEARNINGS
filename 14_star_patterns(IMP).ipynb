```python
#SQUARE PATTERN(WITHOUT SPACE IN END, IT BECOMES RECTANGLE)
#*****
#*****
#*****
#*****
#*****
```

```python
for i in range(1,6):
    for j in range(1,6):
        print("*", end=" ")
    print("")
```

```python
#INCREASING STAR PATTERN
#*
#**
#***
#****
#*****
```

```python
for i in range(1,6):
    for j in range(0,i):
        print("*", end=" ")
    print("")

for m in range(5):
    for n in range(m+1):
        print("*", end=" ")
    print("")
```

```python
#DECREASING STAR PATTERN
#*****
#****
#***
#**
#*
```

```python
for i in range(5, 0, -1):
    for j in range(i, 0, -1):
        print("*", end=" ")
    print("")

for m in range(5):
    for n in range(m, 5):
        print("*", end=" ")
    print("")
```

```python
#RIGHT ANGLED TRIANGLE
#    *
#   **
#  ***
# ****
#*****
```

```python
n = int(input("Enter the number of rows: "))
        
for i in range(n):
    for j in range(i, n-1):
        print(" ", end=" ")
    for k in range(i+1):
        print("*", end= " ")
    print("")
```

```python
#UPSIDE DOWN RIGHT ANGLED TRIANGLE
#*****
# ****
#  ***
#   **
#    *
```

```python
n = int(input("Enter the number of rows: "))

for i in range(n):
    for j in range(i):
        print(" ", end= " ")
    for k in range(i, n):
        print("*", end= " ")
    print("") 
```

```python
#NOT EXACTLY A HILL PATTERN
#    **
#   ****
#  ******
# ********
#**********
```

```python
n = int(input("Enter the number of rows: "))

for i in range(n):
    for j in range(i, n-1):
        print(" ", end=" ")
    for k in range(i+1):
        print("*", end=" ")
    for l in range(i+1):
        print("*", end=" ")
    print("")
```

```python
#HILL PATTERN
#    *
#   ***
#  *****
# *******
#*********
```

```python
n = int(input("Enter the number of rows: "))

for i in range(n):
    for j in range(i, n-1):
        print(" ", end=" ")
    for k in range(i+1):
        print("*", end=" ")
    for l in range(i):
        print("*", end=" ")
    print("")
```

```python
#REVERSE HILL PATTERN
#*********
# *******
#  *****
#   ***
#    *
```

```python
n = int(input("Enter the number of rows: "))

for i in range(n):
    for j in range(i):
        print(" ", end=" ")
    for k in range(i, n):
        print("*", end=" ")
    for l in range(i, n-1):
        print("*", end=" ")
    print("")
```

```python
#DIAMOND PATTREN
#    *
#   ***
#  *****
# *******
#*********
# *******
#  *****
#   ***
#    *
```

```python
n = int(input("Enter the number of rows: "))

for i in range(n):
    for j in range(i, n-1):
        print(" ", end=" ")
    for k in range(i+1):
        print("*", end=" ")
    for l in range(i):
        print("*", end=" ")
    print("")
for t in range(n):
    for m in range(t+1):
        print(" ", end=" ")
    for o in range(t, n-1):
        print("*", end=" ")
    for p in range(t, n-2):
        print("*", end=" ")
    print("")
```
