```python
#TUPLES
#CREATION
#Tuples are created in a similar way like lists!!
T = () #empty tuple
print(T)
print(type(T))

T = tuple()
print(T)
print(type(T))
```

```python
T = (1, 2, 3, 4)
print(T)
T = (1, "Surjo", 2,45, 5 + 3j, True, None, 45) #tuples are also heterogenous!!
print(T)
T = (1, 2, 3, (4, 5)) #multi dimensional tuple
print(T)
```

```python
T = (1)
print(T)
print(type(T)) #this will give int as type because when we declare a single value in tuple it 
#basically gets treated as integer!!

T = ("Surjo")
print(T)
print(type(T)) #HERE TYPE WILL BE CLASS STRIMG!!

#TO AVOID THIS WE HAVE TO ADD AN EXTRA COMMA!!
T = (1,)
print(T)
print(type(T))
```

```python
#DIFFERENCE BETWEEN NORMAL ASSIGNING AND FUNCTION ASSIGNING IN TUPLES!!
T = ("surjo")
print(T)
T = tuple("surjo")
print(T)
```

```python
#WE CAN ALSO CREATE TUPLE FROM A LIST
T = ([1, 2, 3, 4, 5]) #THIS WILL BE TREATED AS A LIST!!
print(T)
print(type(T))

T = tuple([1, 2, 3, 4, 5]) #IF DONE EXPLICIT CONVERSION USING tuple() THEN IT WILL BE TUPLE
print(T)
print(type(T))
```

```python
#ACCESSING IN TUPLE
#THE ACCESSING WORKS SIMILAR LIKE LISTS!!
T= (1, 2, 3, 4, 5)
print(T[0], T[1:3]) #SLICING AND INDEXING WORKS TOO!!
print(T[-2], T[-3:-1])
print(T[-1:-3:-1])
print(T)
```

```python
T = (1, 2, ((3, 4), 5), ((22, 13),(77,54)), 10)
print(T[2][-1], T[3][1][::-1][0])
```

```python
#EDITING, ADDING IN TUPLES
L = [1, 2, 3, 4, 5]
L[1] = 200
print(L) #lists are mutable!!
T = (1, 2, 3, 4, 5)
T[1] = 100
print(T) #tuples are immutable like strings!! sp this will throw error!!
```

```python
#So we also cannot add or edit any value in a tuple, though we can access them!!
```

```python
#DELETING IN TUPLES
T = (1, 2, 3, 4, 5)
del T
print(T)
```

```python
T = (1, 2, 3, 4, 5)
del T[2]
```

```python
#OPERATIONS AND FUNCTIONS
#ARITHMETIC OPERATIONS
T1 = (1, 2, 3, 4) 
T2 = ("Hello", 2.345, True, 21, 5)
print(T1 + T2) #Tuple concatenation!!
print(T1 * 2) #tuple multiplication!!
```

```python
#MEMBERSHIP
T1 = (1, 2, 3, 4, (5, 6))
print(2 in T1)
print(5 in T1)
```

```python
#ITERATION LOOPS 
T = (1, 2, 3, 4, 5, 6, "Surjo", None)
for i in T:
    print(i, end = " ")
print("")
for j in T[6:7]:
    print(j)
```

```python
T = (1, 2, 3, 4, 5, 6, (7, (8, 9)))
print(len(T))
print(min(T)) #this is showing erroe because it can't compare the tuple and int items!!
print(max(T))
```

```python
T = (1, 2, 3, 4, 5)
print(sum(T))
```

```python
T = (1, 3, 5, 7, 2, 6, 9, 4, 8)
print(sorted(T)) #THIS IS NOT A PERMANENT OPERATIOM!!
print(sorted(T, reverse = True)) #THIS ALWAYS GIVES LIST AS A VALUE!!
print(type(sorted(T)))
```
