```python
#GUESS THE NUMBER GAME- (PROJECT)
number=int(input("Guess the correct number between 1 to 100: "))
count=1
correct_num=51 #Here we gave the correct number to be guessed  ourselves!!

while(number!=correct_num):
    if number < correct_num:
        print("Guess a little higher!")
        count+=1
        number=int(input("Guess again!: "))
    else:
        print("Guess a little lower!")
        count+=1
        number=int(input("Guess again!: "))
        
if number == correct_num:
        print("Correct guess, The correct number is",correct_num)
        print("You guessed it in",count,"times")
```
