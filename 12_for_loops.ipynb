```python
#If i am saying a string..like "Kolkata"-> this is a sequence of characters!!
#It is just like range, a sequence of anything..so if want to loop something, 
#it basically needs to go in a sequence form!!
```

```python
#["Kolkata", "Delhi", "Kashmir"] -> so this is a sequence of strings!!
```

```python
for i in range(1, 6):
    print(i)
```

```python
for i in range(10, 0, -1):
    print(i)
```

```python
for i in "Kolkata": #sequence of characters-> string
    print(i)
```

```python
for i in ["Apple", "Banana", "Mango"]: #list of strings
    print("I like", i)
```

```python
for i in [1, 2, 3, 4]: #list
    print(i)
```

```python
for i in (1, 2, 3, 4): #tuple
    print(i)
```

```python
for i in {1, 2, 3, 4}: #set
    print(i)
```
